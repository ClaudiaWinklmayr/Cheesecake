```python
import pickle as p 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
filename = 'panning_'
cols = ['timestamp_franka', 'joint_pos', 'joint_vel', 'endeff_pos', 'timestamp_lidar', 'lidar_data']
df = pd.DataFrame(columns = cols)

for k in range(0, 53): 
    smalldf = p.load(open(filename + str(k) + '.p', 'rb'))
    df = df.append(smalldf, ignore_index = True)
```

```python
p.dump(df, open('alltrajectories_panning_GOOD.p', 'wb'))
```

```python
data = p.load(open('alltrajectories_panning_GOOD.p', 'rb'))
```

```python
plt.plot(np.array(list(data['lidar_data'])))
plt.show()
```
